```python
import gradio as gr
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# here i loaded the data set i got from kaggle
movies = pd.read_csv('/content/data/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/data/tmdb_5000_credits.csv')

# so i dont have the poster path because i ran out of time so its in the code and when improve it in the future i will just change the data set
if 'poster_path' in movies.columns:
    movies = movies[['id', 'title', 'overview', 'poster_path']]
else:
    movies = movies[['id', 'title', 'overview']]

# Fill missing overviews
movies['overview'] = movies['overview'].fillna('')

# Define TF-IDF Vectorizer and calculate similarity > using AI Becuase i had to change some lines
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['overview'])
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# this function will just appear the tittle and recommendations if the movie tittle is correct if not it wont appear
def recommend_movies(title, num_recommendations):
    indices = pd.Series(movies.index, index=movies['title']).drop_duplicates()
    if title not in indices:
        return "Movie not found. Please enter a valid movie title."

    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations + 1]
    movie_indices = [i[0] for i in sim_scores]
    recommendations = movies[['title', 'overview']].iloc[movie_indices]

    movie_details = []
    for _, row in recommendations.iterrows():
        movie_details.append((row['title'], row['overview']))

    return movie_details

# how can we interfer with gardio
def format_output(movie_details):
    output_text = ""
    for title, overview in movie_details:
        output_text += f"**{title}**\n{overview}\n\n"
    return output_text

iface = gr.Interface(
    fn=lambda title, num_recommendations: format_output(recommend_movies(title, num_recommendations)),
    inputs=[
        gr.Textbox(label="Enter a Movie Title", placeholder="Enter a movie title here"),
        gr.Slider(minimum=1, maximum=20, step=1, value=10, label="Number of Recommendations")
    ],
    outputs=gr.Textbox(label="Recommended Movies"),
    title="Movie Recommendation System",
    description="Enter a movie title to get recommendations for similar movies."  # >>>>> this whole line just for the gr play
)
iface.launch(share=True)  #launching the gardio and (share= true) just for it to be shareable
```
